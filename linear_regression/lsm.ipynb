# The least squares method

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)

N = 70 # numbers of experiments
sigma = 3 # standart deviation
# theoretical parameters
k = 0.5
b = 2

f = np.array([k * z + b for z in range(N)]) # line
y = f + np.random.normal(0, sigma, N) # math exp = 0 (center in 0), normal distribution
x = np.array(range(N))

mx = x.sum() / N # math expectation for x
my = y.sum() / N # __ //__ for y
a2 = np.dot(x.T, x) / N # vector ** 2, second start moment
a11 = np.dot(x.T, y) / N

k_model = (a11 - mx * my) / (a2 - mx ** 2)
b_model = my - k * mx

f_model = np.array([k_model * z + b_model for z in range(N)])

sns.scatterplot(x=x, y=y, color='blue',alpha=0.5)
sns.lineplot(f, color='red')
sns.lineplot(f_model, color='black')
plt.grid(True)
```
